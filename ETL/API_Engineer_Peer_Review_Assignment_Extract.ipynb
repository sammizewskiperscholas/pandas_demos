## Extract Data Using an API


### Call the API

<b> Question 1</b> Using the `requests` library call the endpoint given above and save the text, remember the first few characters of the output:


```python
# Write your code here
url = "https://api.apilayer.com/exchangerates_data/latest?base=EUR&apikey=XIwW8eoPh3oczZuBcIUZZc9BnlVb77fa" 
#Make sure to change ******* to your API key.
response = requests.get(url)
data = json.loads(response.text) # find diff b/w load() and loads() ... methods to load json data
data
```

### Save as DataFrame

<b> Question 2</b>  Using the data gathered turn it into a `pandas` dataframe. The dataframe should have the Currency as the index and `Rate` as their columns. Make sure to drop unnecessary columns.


```python
# Turn the data into a dataframere
df = pd.DataFrame(columns=["Currency","Rate"])
df
for k,v in data['rates'].items():
    df = df.append({"Currency":k,"Rate":v}, ignore_index=True)
df
```

```python
# Drop unnescessary columns
df.set_index("Currency", inplace=True, drop=True)
df
```

### Load the Data

Using the dataframe save it as a CSV names `exchange_rates_1.csv`.


```python
# Save the Dataframe
df.to_csv("exchange_rates_1.csv")
```
